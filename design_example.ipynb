```python
# import pya
import pandoc
import pytools_lithography.design.pya as pya
import matplotlib.pyplot as plt
from pytools_lithography.design.KlayoutDesigner import ORIENTATION, Designer, get_template_path
import os

%matplotlib inline
```

# Construct Layout

```python
TEMP_FILENAME = "temp300.gds" # deze naam boeit niet, niet op letten
OUTPUT_FILENAME = "TestVernierDiff.gds" # noem dit whatever

# als je markers ontwerpt laad ze apart in, zonder marker krijg je kruisjes die je kan verwijderen
MARKER_FILENAME = get_template_path("Diamond_Marker.gds") 
```

```python
# Check if the files exist
if os.path.exists(TEMP_FILENAME):
    os.remove(TEMP_FILENAME)
if os.path.exists(OUTPUT_FILENAME):
    print("Output file already exists, please remove it first")
    raise FileExistsError
if not os.path.exists(MARKER_FILENAME):
    print("Marker file not found, please create it first")
    raise FileNotFoundError
```

```python
fig, ax = plt.subplots(figsize = (15, 5))

# Maak een layout via pya.Layout, de resolutie (lekker doen whatever), je aantal layers en verder laten
drawer = Designer(pya.Layout(), 0.001, 3, ax, TEMP_FILENAME, MARKER_FILENAME, OUTPUT_FILENAME)

drawer.bar(-175, 110, 200, 10, 1)
drawer.bar(-175, 130, 200, 10, 2)
drawer.bar(-175, 150, 200, 10, 1)
drawer.bar(-175, 170, 200, 10, 2)
drawer.bar(-175, 190, 200, 10, 1)
drawer.bar(-175, 210, 200, 10, 2)
drawer.bar(-175, 230, 200, 10, 1)
drawer.bar(-175, 250, 200, 10, 2)
drawer.bar(-175, 270, 200, 10, 1)
drawer.bar(-175, 290, 200, 10, 2)

drawer.bar(80, 200, 10, 200, 1)
drawer.bar(100, 200, 10, 200, 2)
drawer.bar(120, 200, 10, 200, 1)
drawer.bar(140, 200, 10, 200, 2)
drawer.bar(160, 200, 10, 200, 1)
drawer.bar(180, 200, 10, 200, 2)
drawer.bar(200, 200, 10, 200, 1)
drawer.bar(220, 200, 10, 200, 2)
drawer.bar(240, 200, 10, 200, 1)
drawer.bar(260, 200, 10, 200, 2)

drawer.vernier(0.0, -700.0, 10.0, 9.5, 5, 1, 2, 700, ORIENTATION.BOTTOM)

drawer.vernier(0.0, -500.0, 10.0, 9.5, 10, 1, 2, 700, ORIENTATION.BOTTOM)
drawer.vernier(-500.0, 0.0, 10.0, 9.5, 10, 1, 2, 700, ORIENTATION.LEFT)

drawer.cross(0,0,10,100,1)
drawer.draw()

drawer.marker(1, (-500, 500), (500, -500), (-500, -500), (500, 500))
```
